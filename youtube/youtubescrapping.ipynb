```python
soup = BeautifulSoup(driver.page_source, "html.parser")
```

```python
#find all links 

links = []
for link in soup.find_all('a'):
    a = link.get('href')
    if("watch" in str(a)):
        links.append("https://www.youtube.com/"+a)
   

#remove duplicates

links = list(dict.fromkeys(links))


```

```python
pdataframe =pd.DataFrame(columns=['view', 'date', 'title', 'link'])
i = 0
list = []

for link in links:
    
    driver.get(link)

    #To allow the page to load

    time.sleep(10)  
    
    soup = BeautifulSoup(driver.page_source, "html.parser",)
    info = get_video_info(soup,link)
    
    list.append(info)
    i = i+1
    print(i)


pdataframe = pd.DataFrame(list)
    
   
```
