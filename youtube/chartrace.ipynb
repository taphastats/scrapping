```python
import pandas as pd
import bar_chart_race as bcr

# Sample Data
data = {
    'Country': ['China', 'India', 'USA', 'UK', 'Germany'],
    '2015': [1376, 1324, 322, 65, 81],
    '2016': [1386, 1339, 324, 66, 82],
    '2017': [1398, 1354, 327, 67, 83]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)
df.set_index('Country', inplace=True)

# Create the bar race animation
bcr.bar_chart_race(
    df=df,
    filename='population_race.mp4',  # Set your desired output filename
    orientation='h',                 # Horizontal bars
    sort='desc',                     # Sort bars descending
    n_bars=10,                       # Show the top 10 bars
    title='Population Bar Race'      # Set a title
)
```
