# YouTube Video Analysis
This notebook contains an analysis of YouTube video statistics for the Baabel series.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
file_path = 'babeel_statistic.xlsx'
df = pd.read_excel(file_path)

# Convert the date column to datetime format
df['date'] = pd.to_datetime(df['date'])

# Calculate summary statistics
summary_stats = df[['views', 'likes', 'comments']].describe()
summary_stats
```

## Summary Statistics

```python
summary_stats
```

## Correlation Matrix

```python
# Correlation matrix
correlation_matrix = df[['views', 'likes', 'comments']].corr()
correlation_matrix
```

## Views, Likes, and Comments Over Time

```python
plt.figure(figsize=(14, 7))

plt.subplot(3, 1, 1)
plt.plot(df['date'], df['views'], label='Views', color='blue')
plt.title('Views Over Time')
plt.xlabel('Date')
plt.ylabel('Views')

plt.subplot(3, 1, 2)
plt.plot(df['date'], df['likes'], label='Likes', color='green')
plt.title('Likes Over Time')
plt.xlabel('Date')
plt.ylabel('Likes')

plt.subplot(3, 1, 3)
plt.plot(df['date'], df['comments'], label='Comments', color='red')
plt.title('Comments Over Time')
plt.xlabel('Date')
plt.ylabel('Comments')

plt.tight_layout()
plt.show()
```

## Scatter Plots

```python
plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.scatter(df['views'], df['likes'], color='blue')
plt.title('Views vs Likes')
plt.xlabel('Views')
plt.ylabel('Likes')

plt.subplot(2, 2, 2)
plt.scatter(df['views'], df['comments'], color='green')
plt.title('Views vs Comments')
plt.xlabel('Views')
plt.ylabel('Comments')

plt.subplot(2, 2, 3)
plt.scatter(df['likes'], df['comments'], color='red')
plt.title('Likes vs Comments')
plt.xlabel('Likes')
plt.ylabel('Comments')

plt.tight_layout()
plt.show()
```

## Histograms

```python
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.hist(df['views'], bins=20, color='blue', edgecolor='black')
plt.title('Distribution of Views')
plt.xlabel('Views')
plt.ylabel('Frequency')

plt.subplot(3, 1, 2)
plt.hist(df['likes'], bins=20, color='green', edgecolor='black')
plt.title('Distribution of Likes')
plt.xlabel('Likes')
plt.ylabel('Frequency')

plt.subplot(3, 1, 3)
plt.hist(df['comments'], bins=20, color='red', edgecolor='black')
plt.title('Distribution of Comments')
plt.xlabel('Comments')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
